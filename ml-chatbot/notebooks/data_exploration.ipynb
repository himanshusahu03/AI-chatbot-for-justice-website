# Data Exploration



```python
def load_data(file_path):
    import pandas as pd
    return pd.read_csv(file_path)

def visualize_response_distribution(data):
    import matplotlib.pyplot as plt
    import seaborn as sns
    plt.figure(figsize=(10, 6))
    sns.countplot(y='response', data=data)
    plt.title('Distribution of Responses')
    plt.xlabel('Count')
    plt.ylabel('Responses')
    plt.show()

def visualize_input_length(data):
    import matplotlib.pyplot as plt
    import seaborn as sns
    data['input_length'] = data['input_text'].apply(lambda x: len(str(x)))
    plt.figure(figsize=(10, 6))
    sns.histplot(data['input_length'], bins=30, kde=True)
    plt.title('Distribution of Input Lengths')
    plt.xlabel('Length of Input')
    plt.ylabel('Frequency')
    plt.show()

# Main execution
data = load_data('../data/input.csv')
visualize_response_distribution(data)
visualize_input_length(data)
```
